# Q1.Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. This task will be done in following steps:

 1.First get the webpage https://www.naukri.com/
        
 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the location” field.

 3.Then click the search button.
    
 4.Then scrape the data for the first 10 jobs results you get.

 5.Finally create a dataframe of the scraped data.

```python
#connect to webdriver
driver = webdriver.Chrome(r"chromedriver.exe")
time.sleep(2)

#to get url
driver.get('https://www.naukri.com/')
```

```python
#finding elements for job search bar
search_job = driver.find_element(By.CLASS_NAME,'suggestor-input ')
search_job.send_keys("Data Analyst")
search_loc = driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
search_loc.send_keys("Bangalore")
```

```python
url = 'https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore'
# to open the webpage through our web drivers
driver.get(url)
```

```python
#  create empty list to save the output
job_titles = []
company_name = []
location = []
experience_required = []

# finding the xpath for job titles
titles_tags = driver.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
titles_tags[:10]

for i in titles_tags:
    title = i.text
    job_titles.append(title)
job_titles[:10]

#finding the xpath for company names
companies_tags = driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
companies_tags[:10]

for i in companies_tags:
    title = i.text
    company_name.append(title)
company_name[:10]

#finding the xpath for locations
location_tags = driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
location_tags[:10]

for i in location_tags:
    title = i.text
    location.append(title)
location[:10]

#finding the xpath for experience
experience_tags = driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
experience_tags[:10]

for i in experience_tags:
    title = i.text
    experience_required.append(title)
experience_required[:10]

#making of dataframe from collected data
jobs = pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company'] = company_name[:10]
jobs['experience'] = experience_required[:10]
jobs['location'] = location[:10]
jobs
```

```python
driver.close
```

```python
# click the search button
search_btn = driver.find_element(By.XPATH,"//div[@class='nav-search-submit nav-sprite']")
search_btn.click()
```

```python
# applying CPU Type filter to “Intel Core i7” filter by checking the respective box
i7_clk=driver.find_elements(By.XPATH,"//span[@class='a-size-base a-color-base']")
for i in i7_clk:
    if i.text == 'Intel Core i7':   
        i.click() # Checking the checkbox to apply filter on the laptop
        break  
```

```python
# creating empty list to scraping data into them

Title=[]
Rating=[]
Price=[]

# extracting all the tags

titles=driver.find_elements(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
for i in titles:
    Title.append(i.text)
    
titles_rat=driver.find_elements(By.XPATH,"//div[@class='a-row a-size-small']/span")
for i in titles_rat:
    Rating.append(i.get_attribute("aria-label"))
        
titles_Price=driver.find_elements(By.XPATH,"//span[@class='a-price']")
for i in titles_Price:
    Price.append(i.text[1:])
```

```python
# creating dataframe for the first 10 laptops data on amazon.com:

amazon =pd.DataFrame({})
amazon['Title']=Title[0:10]
amazon['Rating']=rating[0:10]
amazon['Price']=Price[0:10]

print("The data for the first 10 laptops data on amazon.com:\n")
amazon
```
